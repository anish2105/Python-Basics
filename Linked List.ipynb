```python
#remove duplicates
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

def length(head):
    count = 0
    while head is not None:
        count = count + 1
        head = head.next
    return count

def printnodes(head):
    cur = head
    while cur is not None:
        print(cur.data, end = ' ')
        cur = cur.next
    return 

def takeinput():
    head = None
    mylist = [int(ele) for ele in input("Enter the elements: ").split()]
    
    for j in mylist:
        newnode = Node(j)
        if head is None:
            head = tail = newnode
        else:
            tail.next = newnode
            tail = tail.next
    return head

def remdup(head):
    t1 = head
    t2 = head.next
    while t2 is not None:
        if t1.data != t2.data:
            t1 = t2
            t2 = t2.next
        else:
            while t2 != None and (t1.data == t2.data):
                t2 = t2.next
            t1.next = t2
            t1 = t2
            if t2 != None:
                t2 = t2.next
    return head
            
head = takeinput()
printnodes(head)
print('\n')
head = remdup(head)
printnodes(head)
            
            
            
```

```python
class node:
    def __init__(self,data):
        self.data = data
        self.next = None

class ll:
    def __init__(self):
        self.head = None
        
    def length(self):
        length = 0
        cur = self.head
        while cur is not None:
            length +=1
            cur = cur.next
        return length
    
    def insert(self,data):
        if self.head is None:
            self.head = data
        else:
            lastnode = self.head
            while True:
                if lastnode.next is None:
                    break
                lastnode = lastnode.next
            lastnode.next = data
            
    def insertf(self,data):
        temp = self.head
        self.head = data
        self.head.next = temp
        del temp
        
    def insertm(self,newNode,p):
        if p<0 or p>self.length():
            print("Invalid position")
            return
        if p is 0:
            self.insertf(newNode)
            return
        curposition = 0
        
        previous = None
        cur = self.head
        while True:
            if curposition == p:
                previous.next = newNode
                newNode.next = cur
                break
            previous = cur
            cur = cur.next
            curposition =curposition+1
        
    def printnode(self):
        if self.head is None:
            print("List is empty")
        else:
            cur = self.head
            while True:
                if cur is None:
                    break
                print(cur.data)
                cur = cur.next
                
    def delete(self):
        if self.head == None:
            print("No elements to delete")
            return
        cur = self.head
        while cur.next is not None:
            prev = cur
            cur = cur.next
        prev.next = None
        del cur
    
    def deletef(self,position):
        if position<0 or position>=self.length():
            print("Invalid position")
            return
        
        if position is 0:
            cur = self.head
            self.head = cur.next
            del cur
            return
        
        curr = self.head
        curposition = 0
        while True:
            if curposition == position:
                previous.next = curr.next
                curr.next = None
                del curr
                break
            previous = curr
            curr = curr.next 
            curposition +=1
```
